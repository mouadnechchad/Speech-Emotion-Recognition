## Importing necessary packages

```python
import soundfile #Python module used for reading and writing audio files
import numpy as np
import librosa #is a python package for music and audio analysis
import glob #used to return all file paths that match a specific pattern
import os # to use operating system dependent functionality
import tqdm #used to create progress bars
from sklearn.model_selection import train_test_split # for splitting training and testing 
from sklearn.neural_network import MLPClassifier # multi-layer perceptron model 
from sklearn.metrics import accuracy_score # to measure how good we are
```

## Extract features from the sound file

- First, we will create a function that extracts the determined features from a sound file:
    - **Short-Time Fourier Transform (stft)**: is a Fourier-related transform used to determine the sinusoidal frequency and phase content of local sections of a signal as it changes over time.
    - **Mel-frequency cepstral coefficients (MFCCs)**: are coefficients that collectively make up an MFC, that is a representation of the short-term power spectrum of a sound.
    - **Mel spectrogram (mel)**: is a spectrogram that is converted to a Mel scale. (The Mel Scale is a logarithmic transformation of a signal's frequency.)
    - **Chroma**: is a descriptor, which represents the tonal content of a musical audio signal in a condensed form. 
    - **Spectral Contrast**: is the decibel difference between peaks and valleys in the spectrum.

```python
def get_feature(file_name):
    
        data, sample_rate = librosa.load(file_name)
        stft = np.abs(librosa.stft(data))
        mfccs = np.mean(librosa.feature.mfcc(y=data, sr=sample_rate, n_mfcc=40).T, axis=0)
        mel = np.mean(librosa.feature.melspectrogram(data, sr=sample_rate).T,axis=0)
        chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T,axis=0)
        contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=sample_rate).T,axis=0)
        return mfccs,mel,chroma,contrast
```

## Assigning labels to emotion

- Second, let's create a dictionary where we'll assign a number from 1 to 8 for eight emotions:

```python
# emotions in dataset
list_emotion = {
    "01": "neutral",
    "02": "calm",
    "03": "happy",
    "04": "sad",
    "05": "angry",
    "06": "fearful",
    "07": "disgust",
    "08": "surprised"
}
```

## Loading dataset

- Third, we will create a function to load the dataset and split the data once is loaded at whole into train/test sets:

```python
def load_data(test_size=0.2):
    X,y = [],[]
    for file in tqdm.tqdm(glob.glob("ravdess/Actor_*/*.wav")): #retrieve files/pathnames
        fname = os.path.basename(file) #get basename in the specified path
        emotion =list_emotion[fname.split("-")[2]] #get emotion
        mfccs,mel,chroma,contrast = get_feature(file) #extract features from the file
        
        #assign the extracted features to X and emotions to y
        ext_features = np.hstack([mfccs,mel,chroma,contrast])
        X.append(ext_features)
        y.append(emotion)
    return train_test_split(np.array(X), y, test_size=test_size, random_state=0)
```

```python
X_train,X_test,y_train,y_test = load_data()
```

We see the train/test split shapes:

```python
print("Training shape: ",X_train.shape)
print("Test shape: ",X_test.shape)
```

## MLP Classifier

```python
print("Training the model.....")
clf=MLPClassifier(alpha=0.1, batch_size=300, epsilon=1e-08, hidden_layer_sizes=(500,), learning_rate='adaptive', max_iter=5500).fit(X_train, y_train)
y_pred = clf.predict(X_test)
# calculate the accuracy
accuracy = accuracy_score(y_true=y_test, y_pred=y_pred)
print("Accuracy is: {:.2f}%".format(accuracy*100))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train);
predictions = dtree.predict(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
     
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rforest = RandomForestClassifier(criterion="gini", max_depth=10, max_features="log2", 
                                 max_leaf_nodes = 100, min_samples_leaf = 3, min_samples_split = 20, 
                                 n_estimators= 22000, random_state= 5)
rforest.fit(X_train, y_train)
predictions = rforest.predict(X_test)
print(classification_report(y_test,predictions))
```
